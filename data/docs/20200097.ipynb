문제3) 임의의 정수 리스트를 만드세요.
① input()함수를 사용해서 여러 개의 정수를 한 번에 입력받는다.
② 입력 값을 분리해서 리스트를 만든다. (변수1에 할당)
출력1) print(변수1) #변수1(리스트)를 출력한다.
③ 변수1(리스트)의 원소들을 정수로 수정한다. 반복문 사용.
출력2) print(변수1) #변수1(수정된 리스트)를 출력한다.
<샘플 예제>
‘10 20 30 … 60 50 70’
#①입력값
[’10’, ’20’, ’30’, …, ’60’, ’50’, ’70’]
#②변수1
(입력값을 분리해서 생성)
[10, 20, 30, …, 60, 50, 70] #③변수1
(반복문으로 원소를 수정)

```python
# ① 입력 받기 (공백 없음, 예: 102030)
X= input("정수 목록?")

# ② 2글자씩 분리해서 리스트 만들기 (변수1에 할당)
Y = []
for i in range(0, len(X), 2): # 0부터 시작해서 2씩 건너뜀 (0, 2, 4...)
    Y.append(X[i : i+2]) # i번째부터 i+2번째 앞까지 잘라서 추가

# 출력1) 문자열 리스트 출력
print(Y)

# ③ 반복문을 사용하여 각 원소를 정수(int)로 수정
for i in range(len(Y)):
    Y[i] = int(Y[i])

# 출력2) 수정된 정수 리스트 출력
print(Y)
```

문제4) 올림픽경기의 나라별 금메달 개수가 다음과 같이 저장되어
있다고 하자.
금 = {'Austria': 0, 'Canada': 4, 'China': 5, 'France': 0, 'Germany': 1,
'Japan': 1, 'Korea': 5, 'Norway': 4, 'Russia': 4, 'Viet Nam': 1}
나라이름을 입력받고, 해당 나라의 금메달수를 찾아 출력하라.
3회 반복.
⚫ 안내문구는 ‘나라이름?’
⚫ 목록에 있는 나라 이름이면,
⚫ 해당 나라의 금메달 수를 추출하여 출력한다. (문자열 포맷팅)
⚫ 없는 나라이면,
⚫
‘없는 나라!’라고 출력한다.

```python
GOLD= {'Austria': 0, 'Canada': 4, 'China': 5, 'France': 0, 'Germany': 1,
'Japan': 1, 'Korea': 5, 'Norway': 4, 'Russia': 4, 'Viet Nam': 1}

for i in range(3):
    X=input("나라이름?")
    if X in GOLD:
        print("금메달{}개 ".format(GOLD[X]))
    else:
        print("없는 나라!")
```
